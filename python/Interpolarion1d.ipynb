# Одномерная интерполяция

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import splrep, splev
from Interpolation1d import *

xm = [0,1,2,3,4,5]
ym =[0.1,0.25,0.12,0.62,0.38,0.31]
interpolant=splrep(xm,ym,k=1)
vals=np.linspace(xm[0],xm[len(xm)-1],num=101,endpoint=True)
```

## Линейная интерполяция

```python
res = splev(vals,interpolant)
plt.plot(xm,ym,'bo',label='data')
plt.plot(vals,res,'r--',label='linear')
plt.legend(['data', 'linear'], loc='best')
```

## Интеполяция сплайном Акимы

```python
s = Spline1d.create('A')
aspline = s.build(xm, ym, 2)
resA=aspline.interpolate(vals)
plt.plot(xm,ym,'bo',label='data')
plt.plot(vals,res,'r--',label='linear')
plt.plot(vals,resA,label='akima')
plt.legend(['data', 'linear', 'akima'], loc='best')
```

## Итнерполяция кубическим сплайном

```python
f = splrep(xm, ym, k=3, s=0)
resC = splev(vals,f,der=0)
plt.plot(xm,ym,'bo',label='data')
plt.plot(vals,res,'r--',label='linear')
plt.plot(vals,resA,label='akima')
plt.plot(vals,resC,label='cubic')
plt.legend(['data', 'linear', 'akima','cubic'], loc='best')
```
